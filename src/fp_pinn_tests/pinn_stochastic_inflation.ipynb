# **Fokker-Planck equation for stochastic inflation with PINNs**

## **Inflaton-Langevin equation**
In the slow-roll approximation, the coarse-grained inflaton-Langevin equation is given by

$$ \frac{d\phi_{\rm{cg}}}{d N} = -\frac{V'(\phi_{\rm{cg}})}{3H(\phi_{\rm{cg}})^2}+\frac{H(\phi_{\rm{cg}})}{2\pi}\xi(N) $$
where $\xi(N)$ is a white-noise with 2-pt autocorrelation function of
$$ \langle \xi(N)\xi(N') \rangle=\delta(N-N') $$

We are interested in obtaining the statistics of the stochastic $e$-fold to end of inflation $\mathcal{N}$. The traditional method would be to simply simulate the inflaton-Langevin equation for $10^8 - 10^9$ realizations and generate an empirical distribution. However, such a method is very computationally expensive. More so, it still might not be adequate at resolving the rare event tail of the distribution, which is extremely important for the study of primordial black holes (PBHs).

## **Fokker-Planck equation**
Any white-noise driven process can be converted to an equivalent description. Given a one-dimensional Ito process driven by a Wiener process, given by
$$ dX_t = \mu(X_t, t)dt + \sigma(X_t, t)dW_t $$
with $\mu(X_t, t)$ and $\sigma(X_t, t)$ as drift and diffusion terms, we can derive a deterministic partial differential equation (PDE) can describes the evolution of the probability distribution of the process $p(x,t)$. This is the Fokker-Planck equation
$$ \frac{\partial}{\partial t}p(x,t) = -\frac{\partial}{\partial x}\left[ \mu(x,t)p(x,t) \right]+\frac{1}{2}\frac{\partial^2}{\partial x^2}\left[ \sigma^2(x,t)p(x,t) \right] $$
